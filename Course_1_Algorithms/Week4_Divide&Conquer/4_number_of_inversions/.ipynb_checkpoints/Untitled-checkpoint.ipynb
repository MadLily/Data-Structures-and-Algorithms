```python
# Uses python3
import sys
import random

def partition3(a, l, r):
    #write your code here
    x = a[l]
    j = l;
    kkk = 0
    for i in range(l + 1, r + 1):
        if a[i] < x:
            j += 1
            if kkk!=0:
                a[i], a[j], a[j+kkk] = a[j+kkk], a[i], a[j]
            else:
                a[i], a[j] = a[j], a[i]
        elif a[i] == x:
            kkk+=1
            a[i],a[j+kkk] = a[j+kkk],a[i]
    a[l], a[j] = a[j], a[l]
#    print(a[:j],a[j:j+kkk+1],a[j+kkk+1:])
    return j,kkk

def partition2(a, l, r):
    x = a[l]
    j = l;
    for i in range(l + 1, r + 1):
        if a[i] <= x:
            j += 1
            a[i], a[j] = a[j], a[i]
    a[l], a[j] = a[j], a[l]
    return j


def randomized_quick_sort(a, l, r):
    if l >= r:
        return
    k = random.randint(l, r)
#    print(a[k])
    a[l], a[k] = a[k], a[l]
    #use partition3
#    m = partition2(a, l, r)
#    randomized_quick_sort(a, l, m - 1);
#    randomized_quick_sort(a, m + 1, r);
    m,kk = partition3(a, l, r)
#    print(m,kk)
    randomized_quick_sort(a, l, m - 1);
    randomized_quick_sort(a, m+kk + 1, r);


```

```python
%time gg=randomized_quick_sort([11]*10**5,0,10**5-1)
```
